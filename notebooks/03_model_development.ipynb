```python
import pandas as pd

df_raw = pd.read_csv("../data/raw/korean_stock_data.csv", parse_dates=["timestamp"])
df_processed = pd.read_csv("../data/processed/alpha158_enhanced_features.csv", parse_dates=["timestamp"])

# Check for matching timestamps and closing prices
mismatches = df_processed.merge(df_raw[['timestamp', 'closing_price']], on=['timestamp', 'closing_price'], how='left', indicator=True)
print(mismatches['_merge'].value_counts())

print(mismatches[mismatches['_merge'] != 'both'])
```
